```python
def get_relevance_idxs(df, thresh):
    return df[df['mIndex'] > thresh][['mIndex', 'sign']]


def filter_by_relevance_idxs(df, col='mIndex', thresh=0.5):
    return df[df[col] > thresh]


def get_date_range(date, d_range):
    date_list = []
    date_list.append(date)
    for day in range(1, d_range+1):
        b_date = (date - datetime.timedelta(days = day))
        a_date = (date + datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.append(a_date)
        date_list.sort()
    return date_list


# The cast needs to be handled more elegantly
def filter_by_previous_dates(df, date, date_range):
    mindate = str((pd.to_datetime(date, format='%Y/%m/%d', errors='coerce') - pd.to_timedelta(date_range, unit='days')).date())
    lowerbound = df[df.index > mindate]    
    return lowerbound[lowerbound.index < date]


def get_previous_dates(date, date_range):
    date_list = []
    date_list.append(date)
    for day in range(1, date_range+1):
        b_date = (date - datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.sort()
    return date_list


# def get_volumes_in_range(df, days_range):
#     spikes = get_days_left(iris_join, 0.5)
#     days_range = timedelta(days=days_range)
#     for date in spikes['date_index']:


def filter_by_date_range(dataset, mindate, maxdate):
    lowerbound = dataset[dataset.index > mindate]    
    return lowerbound[lowerbound.index < maxdate]
```
