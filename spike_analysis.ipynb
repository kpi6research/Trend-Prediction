```python
def get_relevance_idxs(df, thresh):
    return df[df['mIndex'] > thresh][['mIndex', 'sign']]


def filter_by_relevance_idxs(df, col='mIndex', thresh=0.5):
    return df[df[col] > thresh]


def get_date_range(date, d_range):
    date_list = []
    date_list.append(date)
    for day in range(1, d_range+1):
        b_date = (date - datetime.timedelta(days = day))
        a_date = (date + datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.append(a_date)
        date_list.sort()
    return date_list


# The cast needs to be handled more elegantly
def filter_by_previous_dates(df, date, date_range, col=None, formatted=False):
    formatted_date = pd.to_datetime(date, format='%Y/%m/%d', errors='coerce')
    mindate = (formatted_date - pd.to_timedelta(date_range, unit='days')).date()
    if(not formatted):
        mindate = str(mindate)
    if (col is None):
        lowerbound = df[df.index > mindate]    
        return lowerbound[lowerbound.index < date]
    else:
        lowerbound = df[df[col] > mindate]  
        return lowerbound[lowerbound[col] < formatted_date]


def get_previous_dates(date, date_range):
    date_list = []
    date_list.append(date)
    for day in range(1, date_range+1):
        b_date = (date - datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.sort()
    return date_list


# def get_volumes_in_range(df, days_range):
#     spikes = get_days_left(iris_join, 0.5)
#     days_range = timedelta(days=days_range)
#     for date in spikes['date_index']:


def filter_by_date_range(dataset, mindate, maxdate, col=None):
    formatted_mindate = pd.to_datetime(mindate, format='%Y/%m/%d', errors='coerce').date()
    formatted_maxdate = pd.to_datetime(maxdate, format='%Y/%m/%d', errors='coerce').date()
    if (col is None):
        lowerbound = dataset[dataset.index > formatted_mindate]    
        return lowerbound[lowerbound.index < formatted_maxdate]
    else:
        lowerbound = dataset[dataset[col] > formatted_mindate]    
        return lowerbound[lowerbound[col] < formatted_maxdate]


def convert_datetime_col_to_iso_format(df, col='date_index'):
    df[col] = df[col].apply(lambda x: pd.to_datetime(x, format='%Y/%m/%d', errors='coerce').date())
```

```python
def get_all_spikes_content(df_spikes, df_spike_tweets, dates_delta=7):
    spike_dates = {'spike_date': [], 'id': [], 'auth_username': []}
    spike_tweets_df = {}
    for date in df_spikes.index:
        spike_dates['spike_date'].append(date)
        spike_tweets_df = filter_by_previous_dates(df_spike_tweets, date, dates_delta, 'date_index', formatted=True)
        spike_tweets = {'id': [], 'auth_username': []}
        for tweet in spike_tweets_df:
            spike_tweets['id'].append(tweet['id'])
            spike_tweets['auth_username'].append(tweet['user_username'])
        spike_dates['id', 'auth_username'].append(spike_tweets)
    return pd.DataFrame(spike_dates, index=spike_dates)
```
