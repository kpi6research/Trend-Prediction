```python
def get_days(df, thresh):
    return df[df['mIndex'] > thresh][['mIndex', 'sign']]


# Same as get_days, but on the left or right side of the merged dataset
def get_days_left(df, thresh):
    df_copy = df.copy()
    return df_copy[df_copy['mIndex_x'] > thresh][['mIndex_x', 'sign_x']]


def get_days_right(df, thresh):
    return df[df['mIndex_y'] > thresh][['mIndex_y', 'sign_y']]


def get_date_range(date, d_range):
    date_list = []
    date_list.append(date)
    for day in range(1, d_range+1):
        b_date = (date - datetime.timedelta(days = day))
        a_date = (date + datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.append(a_date)
        date_list.sort()
    return date_list


def get_dates_until(date, d_range):
    date_list = []
    for day in range(1, d_range+1):
        b_date = (date - datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.sort()
    return date_list

# def get_volumes_in_range(df, days_range):
#     spikes = get_days_left(iris_join, 0.5)
#     days_range = timedelta(days=days_range)
#     for date in spikes['date_index']:

def get_data_between(dataset, mindate, maxdate):
    lowerbound = dataset[dataset.date_index > mindate]    
    return lowerbound[lowerbound.date_index < maxdate]
```

```python
date = pd.to_datetime('2019-12-7 00:00:00+00:00')
date_range = get_dates_until(date, 7)
date_range_idx = pd.to_datetime(date_range, utc=True)
date_range_idx = pd.DataFrame(date_range_idx, columns=['date'])
date_range_idx
```
