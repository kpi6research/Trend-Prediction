```python
def get_relevance_idxs(df, thresh):
    return df[df['mIndex'] > thresh][['mIndex', 'sign']]


def filter_by_relevance_idxs(df, col='mIndex', thresh=0.5):
    return df[df[col] > thresh]


def get_date_range(date, d_range):
    date_list = []
    date_list.append(date)
    for day in range(1, d_range+1):
        b_date = (date - datetime.timedelta(days = day))
        a_date = (date + datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.append(a_date)
        date_list.sort()
    return date_list


def filter_by_previous_dates(df, date, date_range, col=None, include_date=False):
    df = df.copy()
    formatted_date = pd.to_datetime(date, format='%Y/%m/%d', errors='coerce')
    # formatted_date = (formatted_date - pd.to_timedelta(1, unit='days'))            # Needed to compare days by String
    mindate = (formatted_date - pd.to_timedelta(date_range, unit='days')).date()    
    if include_date is True:                                                         # Includes the signature date into the range
        # mindate += timedelta(days=1)
        formatted_date += timedelta(days=1)
    if (col is None):
        df = df.reset_index()
        lowerbound = df[df.iloc[:, -1] >= mindate]
        lowerbound = lowerbound[lowerbound.iloc[:, -1] < formatted_date]
        lowerbound = lowerbound.set_index(lowerbound.iloc[:, -1]).drop(lowerbound.columns[-1], axis=1)
        return lowerbound
    else:
        lowerbound = df[df[col] >= mindate]  
        return lowerbound[lowerbound[col] < formatted_date]


def filter_by_following_dates(df, date, date_range, col=None, include_date=False):
    df = df.copy()
    formatted_date = pd.to_datetime(date, format='%Y/%m/%d', errors='coerce')
    maxdate = (formatted_date + pd.to_timedelta(date_range, unit='days')).date()
    if include_date is True:                                                         # Includes the signature date into the range
        # maxdate -= timedelta(days=1)  
        formatted_date -= timedelta(days=1)  
    if (col is None):
        df = df.reset_index()
        lowerbound = df[df.iloc[:, -1] > formatted_date]
        lowerbound = lowerbound[lowerbound.iloc[:, -1] <= maxdate]
        lowerbound = lowerbound.set_index(lowerbound.iloc[:, -1]).drop(lowerbound.columns[-1], axis=1)
        return lowerbound
    else:
        lowerbound = df[df[col] > formatted_date]  
        return lowerbound[lowerbound[col] <= maxdate]


def get_previous_dates(date, date_range):
    date_list = []
    date_list.append(date)
    for day in range(1, date_range+1):
        b_date = (date - datetime.timedelta(days = day))
        date_list.append(b_date)
        date_list.sort()
    return date_list


def filter_by_date_range(dataset, mindate, maxdate, col=None):
    formatted_mindate = pd.to_datetime(mindate, format='%Y/%m/%d', errors='coerce').date()
    formatted_maxdate = pd.to_datetime(maxdate, format='%Y/%m/%d', errors='coerce').date()
    if (col is None):
        lowerbound = dataset[dataset.index > formatted_mindate]    
        return lowerbound[lowerbound.index <= formatted_maxdate]
    else:
        lowerbound = dataset[dataset[col] > formatted_mindate]    
        return lowerbound[lowerbound[col] <= formatted_maxdate]


def convert_datetime_col_to_iso_format(df, col=None):
    df = df.copy()
    if (not col):
        df = df.reset_index()
        df['date_index'] = df['date_index'].apply(lambda x: pd.to_datetime(x, format='%Y/%m/%d', errors='coerce').date())
        df = df.set_index(keys=['date_index'])
        return df
    else:
        df['date_index'] = df['date_index'].apply(lambda x: pd.to_datetime(x, format='%Y/%m/%d', errors='coerce').date())
        return df
```

```python
# We need to explicitate each spike - tweet relashionship over the course of a chosen time delta. 
def get_all_spikes_content(df_spikes, 
                           df_spike_tweets, 
                           dates_delta=7,
                           include_date=False):
    spike_tweets_lst = []
    for i, row in df_spikes.iterrows():
        date = row['date_index']
        spike_tweets_df = filter_by_previous_dates(df_spike_tweets, 
                                                   date, 
                                                   dates_delta, 
                                                   'date_index',
                                                   include_date)
        spike_tweets_df['spike_id'] = row['index']
        spike_tweets_lst.append(spike_tweets_df)
    return pd.concat(spike_tweets_lst)
```

```python
def get_formatted_date(date):
    return pd.to_datetime(date, format='%Y/%m/%d', errors='coerce').date()

def collapse_consecutive_spike_days(df):
    output = df.copy()
    collapsed_spikes = 0
    formatted_date = get_formatted_date(df['spike_window_date'].iloc[0])
    prev_date = formatted_date
    spike_starting_date = formatted_date
    spike_ending_date = formatted_date
    output['spike_starting_date'] = spike_starting_date
    output['spike_ending_date'] = spike_ending_date
    last_spike_start_row_idx = -1
    prev_spike = df['spike_id'].iloc[0]
    
    for i, row in df.iterrows():
        formatted_date = get_formatted_date(row['spike_window_date'])
        curr_prev_date = formatted_date - pd.to_timedelta(1, unit='days')
        if(curr_prev_date == prev_date):
            collapsed_spikes += 1
        elif(formatted_date != prev_date):
            spike_starting_date = formatted_date
        
        prev_date = formatted_date
        curr_spike = row['spike_id'] - collapsed_spikes
        # prev_spike = prev_spike - collapsed_spikes
        output['spike_id'].iloc[i] = curr_spike
        output['spike_starting_date'].iloc[i] = spike_starting_date
        output['spike_ending_date'].iloc[i] = formatted_date
        try:
            formatted_ending_date = get_formatted_date(df['spike_window_date'].iloc[i+1])       # execute the next block only if there is a succeeding tuple
            if (formatted_ending_date > (formatted_date + pd.to_timedelta(1, unit='days'))):    # if the next date has a timedelta superior to 1 from the examined one, it means that it belongs to another spike
                if (curr_spike==prev_spike):                                                    # update the ending date only if the spike isn't changed
                    for j in range(last_spike_start_row_idx, i):
                        output['spike_ending_date'].iloc[j] = formatted_date                    
                last_spike_start_row_idx = i+1     
        except:
            print('There are not next lines in the DataFrame')
        else:
            last_row_idx = i
        prev_spike = curr_spike

        if (curr_spike==prev_spike):
            for k in range(last_spike_start_row_idx, last_row_idx + 1):
                output['spike_ending_date'].iloc[k] = formatted_date
    return output
```

```python
def get_tweets_in_date_range(df, 
                             date_series,
                             range_type, 
                             date_range=7, 
                             range_start=7,
                             range_end=7,
                             include_date=False):
    
    range_types = ['until', 'beyond', 'in_range']
    
    if range_type not in range_types:
        raise Exception('You must specify a date_range_type. The accepted values are [until, beyond, in_range]')
    
    output = []

    if(range_type==range_types[0]):
        for i, row in date_series.iterrows():
            temp_df = filter_by_previous_dates(df, row['spike_starting_date'], date_range, 'tweet_date', include_date)
            output.append(temp_df)
    elif(range_type==range_types[1]):
        for i, row in date_series.iterrows():
            temp_df = filter_by_following_dates(df, row['spike_starting_date'], date_range, 'tweet_date', include_date)
            output.append(temp_df)
    elif(range_type==range_types[2]):
        for i, row in date_series.iterrows():
            start_date = row['spike_starting_date'] - timedelta(days=range_start)
            end_date = row['spike_starting_date'] + timedelta(days=range_end)     
            start_date = row['spike_starting_date'] - timedelta(days=range_start)
            end_date = row['spike_starting_date'] + timedelta(days=range_end)
            temp_df = filter_by_date_range(df, start_date, end_date, 'tweet_date')
            output.append(temp_df)
    return pd.concat(output)
```

```python
# Adds a new column counting the length of each spike in days to the input dataframe
def get_spikes_days_length(df, id_col='spike_id', start_date_col='spike_starting_date', end_date_col='spike_ending_date'):
    output = df.copy()
    days_deltas = []
    for i, row in output.iterrows():
        days_delta = ((row[end_date_col] - row[start_date_col]) + timedelta(days=1))
        days_deltas.append(days_delta.days)
    output['spike_length'] = days_deltas
    return output
```
